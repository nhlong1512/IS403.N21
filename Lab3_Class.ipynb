```python
import pandas as pd 
import numpy as np
home_value = pd.read_csv("home_value.csv", sep=";")
home_value
```

```python
x, y = np.array(home_value[["Square Feet", "House Age"]]).reshape((-1,2)), np.array(home_value["Market Value"]).reshape((-1,1))
```

```python
%pip install sklearn
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
print("Hệ số chẵn: ", model.intercept_)
print("Hệ số phụ thuộc: ", model.coef_)
print("Điểm R Square: ", model.score(x, y))
```
